```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('timing_results.csv')

# Plotting
plt.figure(figsize=(10, 6))

# Plot insertion times
plt.plot(df['n'], df['PerfectHashingInsertTime'] / df['n'], label='Perfect Hashing Insertion Time')
plt.plot(df['n'], df['HashingWithChainingInsertTime'] / df['n'], label='Hashing with Chaining Insertion Time')
plt.plot(df['n'], df['RedBlackTreeInsertTime'] / df['n'], label='Red-Black Tree Insertion Time')

# Plot querying times
plt.plot(df['n'], df['PerfectHashingQueryTime'] / df['n'], linestyle='--', label='Perfect Hashing Querying Time')
plt.plot(df['n'], df['HashingWithChainingQueryTime'] / df['n'], linestyle='--', label='Hashing with Chaining Querying Time')
plt.plot(df['n'], df['RedBlackTreeQueryTime'] / df['n'], linestyle='--', label='Red-Black Tree Querying Time')

plt.xlabel('Input Size (n)')
plt.ylabel('Average Operation Time')
plt.title('Comparison of Data Structure Performance')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('data_structure_performance.png')

# Show the plot
plt.show()
```
